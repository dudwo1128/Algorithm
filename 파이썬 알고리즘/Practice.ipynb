```python
d = [1,4,9,16,25,36,49,64,81]

def bin_search(a,n):
    start = 0
    end = len(a) - 1
    
    while start <= end:
        mid = (start+end)//2
        if n == a[mid]:
            return mid
        elif a[mid] < n :
            start = mid + 1
        else:
            end = mid - 1
    return -1

bin_search(d,36)
```

```python
def palindrone(a):
    qu = []
    st = []
    
    for i in a:
        if i.isalpha():
            qu.append(i)
            st.append(i)
    while qu:
        if qu.pop(0) != st.pop():
            return False
    return True

palindrone("Good")
```

```python
name1 = ["Lee","Tom","Jerry","Mike","Tom","Mike","Jerry","Kim","Lee"]
def find_same_name(a):
    name_dict={}
    for name in a:
        if name in name_dict:
            name_dict[name] += 1
        else:
            name_dict[name] = 1
    
    result = []
    for name in name_dict:
        if name_dict[name] > 1:
            result.append(name)
    return result

find_same_name(name1)
```
