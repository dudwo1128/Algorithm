#### Tom의 경우

2. Tom 친구 중에 Jerry 있으므로 Jerry 출력
3. 끝

#### Summer의 경우
1. Summer 본인 출력
2. Summer 친구 3명 출력
3. Summer 친구들의 친구 출력.... but 친구 너무 많음

## 해결 방법 : 2개의 메모 필요!

1. 앞으로 처리할 사람들(qu)
    - 친구의 친구들을 모두 메모
    - 처리할 때마다 목록에서 제거


2. 이미 추가 된 사람들(done)
    - 한명이 여러번 나온 경우
    - 중복되지 않도록

```python
def print_all_friends(g,start):
    qu = []
    done = set()
    
    qu.append(start)
    done.add(start)
    
    while qu:
        p = qu.pop(0)
        print(p)
        for x in g[p]:
            if x not in done:
                qu.append(x)
                done.add(x)

print_all_friends(fr_info,'Kim')
```
